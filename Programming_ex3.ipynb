```python
import re
import numpy as np
import json

n_word = 48
n_char = 6

word_list = []
char_list = []
others = []
features = open('features.txt', 'r')
for i, line in enumerate(features):
    word = line.split(':')[0]
    if i > 32:
        if i < (33+n_word):
            word = re.sub('word_freq_', '', word)
            word_list.append(word)
        elif i < (33+n_word+n_char):
            char = re.sub('char_freq_', '', word)
            char_list.append(char)
        else:
            others.append(word)
print ('\n word list = \n', word_list)
print('\n char list = \n', char_list)
print('\n other features = \n', others)
features.close()
```

```python
#Create features as a dictionary object

features_dict = {'word_freq_'+key:None for key in word_list}
chars = {'char_freq_'+key:None for key in char_list}
others = {key:None for key in others}
features_dict.update(chars)
features_dict.update(others)
```

```python
def extract_features(filename):
    print('\n\n'+ filename+'\n\n')
    
    file = open(filename, 'r')
    content = file.read()
    total_words = len(re.findall(r'\w+', content))
    
    texts = content.lower() 
    for word in word_list:
        match_pattern = re.findall(' '+word+' ', texts)
        features_dict['word_freq_'+word] = len(match_pattern)*100/total_words

    for char in char_list:
        if char in ['(', '{', '[', '$']:
            match_pattern = re.findall('\\'+char, texts)
        else:
            match_pattern = re.findall(char, texts)
        features_dict['char_freq_'+char] = len(match_pattern)


    caps = re.findall(r"[A-Z]+", content)

    max_length = len(max(caps, key = len))
    total_no = sum(len(x) for x in caps)
    average_length = total_no/len(caps)

    features_dict['capital_run_length_longest'] = max_length
    features_dict['capital_run_length_total'] = total_no
    features_dict['capital_run_length_average'] = average_length
    
    file.close()
    
    print(print(json.dumps(features_dict, indent = 4)))
    return features_dict
```

```python
files = ['spam_or_no_spam.txt', 'spam_or_no_spam_2.txt', 'spam_or_no_spam_3.txt']

features_arr = []

for file in files:
    features_arr.append(extract_features(file))
```
